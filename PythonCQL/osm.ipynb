```python
import pandas as pd
import geopandas as gpd

df = pd.read_csv('NodosOSM.csv')
df.drop(columns=['elementid(n)'], inplace=True)
df = df.rename(columns={'n.node_id': 'id', 'n.longitude': 'longitude', 'n.latitude' :'latitude'})
df
```

```python
centro = df.loc[df['id'] == 1394426259]
centro
```

```python
import geopandas as gpd
from shapely.geometry import Point
from geopy.distance import geodesic

# Crear un DataFrame con tus datos (asumiremos que df está definido previamente)
df = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['longitude'], df['latitude']))

# Definir el punto arbitrario
punto_arbitrario = Point(20.63121, -103.425199)

# Calcular la distancia entre cada punto y el punto arbitrario, y agregarla como una columna en el DataFrame
df['distancia'] = df['geometry'].apply(lambda x: geodesic((x.x, x.y), (punto_arbitrario.x, punto_arbitrario.y)).meters)

# Filtrar los puntos que están a una distancia menor o igual a un kilómetro
puntos_cercanos = df[df['distancia'] <= 1000]  # 1000 metros = 1 kilómetro

# Mostrar los puntos cercanos
puntos_cercanos.drop(columns=['distancia', 'geometry'], inplace=True)
puntos_cercanos

```

```python
puntos_cercanos.to_csv('kmred.csv', index=False, header=False)
```
